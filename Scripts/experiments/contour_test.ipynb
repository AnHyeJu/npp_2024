```python
import numpy as np
import pandas as pd
import xarray as xr

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
#   
lon = np.linspace(-180, 180, 100)
lat = np.linspace(-90, 90, 50)
lon2d, lat2d = np.meshgrid(lon, lat)
data = np.sin(np.radians(lat2d)) * np.cos(np.radians(lon2d))

#   np.nan 
data[10:20, 30:40] = np.nan

# 
fig, ax = plt.subplots(subplot_kw={'projection': ccrs.Robinson(central_longitude=180)})
contour = ax.contourf(lon, lat, data, transform=ccrs.PlateCarree(), cmap='viridis')

# coastlines 
ax.coastlines()

#  
cbar = plt.colorbar(contour, ax=ax, orientation='vertical')
cbar.set_label('Data values')

plt.show()
```
